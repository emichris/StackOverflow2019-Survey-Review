```python
pay_df = df[df["Employment"]=="Employed full-time"]
pay_df.dropna(subset=["ConvertedComp"], axis=0, inplace=True)
pay_df.groupby("Country").count()["ConvertedComp"].sort_values(ascending=False).nlargest(20).plot(kind="bar", figsize=(20,5), rot=45);
```

```python
# How does the age you first code affect your satisfaction as a software engineer
```

```python
pay_df.CareerSat.unique()
```

```python
pay_df = df[df["Employment"]=="Employed full-time"]
pay_df.dropna(subset=["Age1stCode"], axis=0, inplace=True)
pay_df.dropna(subset=["CareerSat"], axis=0, inplace=True)
# Convert the categories into a numeric scale from -2 to +2

cat_sat = {'Very dissatisfied': -2, 'Slightly satisfied': -1, 
           'Neither satisfied nor dissatisfied': 0, 'Very satisfied': 1,'Slightly dissatisfied': 2 }

sat_num = lambda x: cat_sat[x]
pay_df["num_CareerSat"] = pay_df.CareerSat.apply(sat_num)

pay_df.groupby("Age1stCode").mean()["num_CareerSat"].sort_values(ascending=False).nlargest(20).plot(kind="bar", figsize=(20,5), rot=45);
```

```python
# If you are breaking into software engineering at an older age, this must be for fun! 
```

```python
pay_df.groupby("Age1stCode").count()["num_CareerSat"].sort_values(ascending=False).nlargest(20).plot(kind="bar", figsize=(20,5), rot=45);
```

```python
# Many people are starting to program in their teenage years. 
# The introduction of online programming courses and adaptation of computer science into middle school and high school curriculum is paying off.
```

```python
# Get average pay by years of professional programming in the United States
def years2num(x):
    if x=="Less than 1 year":
        return 0
    else:
        if x=="More than 50 years": 
            return 51
        else: 
            return int(x)

pay_df.dropna(subset=["YearsCodePro"], axis=0, inplace=True)
pay_df["num_YearsCodePro"] = pay_df.YearsCodePro.apply(years2num)
df_ = pay_df.groupby("num_YearsCodePro").mean()["ConvertedComp"].sort_index().plot(kind="bar", figsize=(20,5), rot=45);
# For beginners what country has the best compensation-this is hard because of the local economy of the country
df_
```

```python
# There seems to be little correlation between the years of programming and pay. 
# However there sees to be a consistent growth from years 2 to 27. 
```
