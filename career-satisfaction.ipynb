```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline

df = pd.read_csv("./data/survey_results_public.csv")
pd.set_option('max_columns', None) # set pandas to display all columns
```

```python
df = df_usa[["MainBranch", "Hobbyist", "OpenSourcer", "OpenSource",
        "Employment", "Student", "EdLevel", "UndergradMajor", "EduOther", 
        "OrgSize", "DevType", "YearsCode", "Age1stCode", "YearsCodePro", "CareerSat", 
        "JobSat", "MgrIdiot", "MgrMoney", "MgrWant", "JobSeek", "LastHireDate", "CompTotal"]]
df.head()
```

```python
# Observe the distribution of the dataset based on educational level

counts = df.groupby('EdLevel').size().sort_values(ascending=False).reset_index(name='Count')
counts["Percentage"] = counts['Count'] / np.sum(counts['Count'])
counts
```

## Group the data into highschool / post high school

In other to see how people with only Hgh School degree compare in the survey engineering market, they are grouped with those with some college and compared to those who have a bachelor's, Master's, Doctorate or Professional Degree

```python
HighSchool_SomeCollege = df[df.EdLevel.isin(['Some college/university study without earning a degree', 'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)'])]
AboveHighSchool = df[df["EdLevel"].isin(['Bachelor’s degree (BA, BS, B.Eng., etc.)',
       'Master’s degree (MA, MS, M.Eng., MBA, etc.)', 'Other doctoral degree (Ph.D, Ed.D., etc.)',
        'Professional degree (JD, MD, etc.)'])]
HighSchool_SomeCollege.head(5)
```

```python
print(HighSchool_SomeCollege.shape)
print(AboveHighSchool.shape)
```

```python
# How do they compare in career satisfaction?
fig, axes = plt.subplots(ncols=2, figsize=(15, 4), dpi=100)

HighSchool_SomeCollege.CareerSat.value_counts(normalize=True).plot(kind="bar", ax=axes[0], rot=45)
AboveHighSchool.CareerSat.value_counts(normalize=True).plot(kind="bar", ax=axes[1], rot=45)

axes[0].set_ylabel('Career Satisfaction', size=15, labelpad=20)
axes[0].tick_params(axis='x', labelsize=12)
axes[0].tick_params(axis='y', labelsize=12)
axes[1].tick_params(axis='x', labelsize=12)
axes[1].tick_params(axis='y', labelsize=12)

axes[0].set_title('High School/Some College', fontsize=14)
axes[1].set_title('Post High School Degree', fontsize=14)

plt.show()

```

## Going further to see how people are staying competitive in the field

```python
# What other forms of eduction are people getting
studyH = HighSchool_SomeCollege['EduOther'].value_counts().reset_index().rename(columns={'index': 'method', 'EduOther' : 'count'})
studyP = AboveHighSchool['EduOther'].value_counts().reset_index().rename(columns={'index': 'method', 'EduOther' : 'count'})
print(studyP.values[2])
```

```python
# Some of the responses has multiple options combined in one row.
# this function will assist in splitting the repsonses into individual methods

def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]
    
    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up 
    '''
    new_df = defaultdict(int)
    for val in look_for:
        for idx in range(df.shape[0]):
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])   
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df
```

```python
possible_vals = ['Taken an online course in programming or software development (e.g. a MOOC)',
        'Participated in a full-time developer training program or bootcamp',
        'Taken a part-time in-person course in programming or software development',
        'Completed an industry certification program (e.g. MCPD)', 
        'Received on-the-job training in software development',
        'Taught yourself a new language, framework, or tool without taking a formal course',
        'Participated in online coding competitions (e.g. HackerRank, CodeChef, TopCoder)',
        'Participated in a hackathon', 'Contributed to open source software',
        'None of these']
                 
```

```python
# What percentage is taking class online? How are they making it work?

studyH_df = total_count(studyH, 'method', 'count', possible_vals)
studyP_df = total_count(studyP, 'method', 'count', possible_vals)
studyH_df["Percentage"] = studyH_df['count']/np.sum(studyH_df['count'])
studyP_df["Percentage"] = studyP_df['count']/np.sum(studyP_df['count'])


# How do they compare in non traditional educational methods?
fig, axes = plt.subplots(nrows=2, figsize=(25, 10), dpi=100)

studyH_df.plot(kind="barh", x='method', y='Percentage', ax=axes[0], rot=0)
studyP_df.plot(kind="barh", x='method', y='Percentage', ax=axes[1], rot=0)

axes[0].set_ylabel('Other forms of education', size=15, labelpad=20)
axes[0].tick_params(axis='x', labelsize=14)
axes[0].tick_params(axis='y', labelsize=14)
axes[1].tick_params(axis='x', labelsize=14)
axes[1].tick_params(axis='y', labelsize=14)

axes[0].set_title('High School/Some College', fontsize=15)
axes[1].set_title('Post High School Degree', fontsize=15)

plt.show()

```
